```python
#my_id_list = ["34614328", "34281357","34310793","34849657","34916207"]
my_id_list = ["34614328"]
data_summary= pd.DataFrame()

i = 0

for pmid in my_id_list:
  file_name = "article_summary_"+pmid+".npy"

  data = np.load(file_name,allow_pickle='TRUE').item()
  data_summary = data_summary.append(data, ignore_index=True)
  i=i+1

  print(i, file_name)
#print(data['conclusion'])
```

```python
data_summary.head()
```

```python
intro1 = data_summary['intro'].iloc[0]
len(intro1)
```

```python
#intro = data_summary['intro'].str.cat()
#len(intro)
full_text = data_summary['method'].iloc[0] + data_summary['results'].iloc[0] + data_summary['conclusion'].iloc[0]
len(full_text)
```

```python
tokenizer(full_text, max_length=20000, padding="max_length", truncation=True, return_tensors="pt")
```

```python
inputs_dict = tokenizer(full_text, max_length=20000, padding="max_length", truncation=True, return_tensors="pt")
input_ids = inputs_dict.input_ids
attention_mask = inputs_dict.attention_mask
```

```python
output_ids = model.generate(input_ids, attention_mask=attention_mask, max_length=512, num_beams=2)
```

```python
predicted_intro = tokenizer.batch_decode(output_ids, skip_special_tokens=True)
predicted_intro
```
